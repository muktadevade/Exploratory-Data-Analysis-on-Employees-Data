## **Employees Attrition Rate : Exploratory Data Analysis**

## **Problem Statement**

*	People Charm a growing company is facing a high attrition rate among their employees which in turn affects their business due to lack of expertise and experience.

*	There HR department is assigned the task to reduce the attrition rate by retaining employees who are about to churn out.


*	They need to recommend special plans or strategies which will help to them to retain their employees which in turn will help them to grow bigger as a company.





Importing Essential

```python
# Importing Libraries

# for data manipulation or analysis
import pandas as pd

# for numeric calculation
import numpy as np

# for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing Dataset

```python
# Loading Dataset
dataset = pd.read_csv('/content/employees.csv')
```

```python
# Shape of Dataset
dataset.shape
```

```python
# Looking at the Dataset of First 5 Rows
dataset.head(5)
```

```python
# Looking the Last 5 Rows of Dataset
dataset.tail(5)
```

```python
# Descriptive Statistics : It provides purely descriptive information about the dataset. This information includes statistics
#                          that summarize the central tendency of the variable,their dispersion, the presence of empty values and their shape.

dataset.describe()
```

```python
# Information about the data type, non-null values and memory usage.

dataset.info()
```

```python
  # To check for duplicate rows we can use .isduplicated().sum()— this will print us the number of duplicated rows in our dataset

  dataset.duplicated().sum()
```

```python
# The drop_duplicates() method removes duplicate rows.
# Use the subset parameter if only some specified columns should be considered when looking for duplicates.
# dataframe.drop_duplicates(subset, keep, inplace, ignore_index)

New_Dataset = dataset.drop_duplicates()
```

```python
# Dataset After Removing Duplicate Rows

New_Dataset
```

```python
# Categorical variables
# To Count The Values in the dataset  ----> 0=not left , 1=left

New_Dataset.left.value_counts()
```

```python
#  Expressing the data as a percentage by passing normalize = True

New_Dataset.left.value_counts(normalize=True)
```

Visualization of Data

```python
# Visualizing the count of left employee and remain(leave=1 , other=0)
New_Dataset.left.value_counts().plot(kind="bar")
plt.title("Value counts of the Left Employee")
plt.xlabel("Left")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
# To visualize the relationship between satisfactorylevel & left no. of employee we assign the variable for each column

a = New_Dataset.satisfactoryLevel
b = New_Dataset.left
```

```python
# Plotting histogram to see relationship between left , satisfactory level

plt.hist(a)
plt.hist(b)
plt.title("satisfactorylevel compairing with left")
plt.xlabel("staisfactory_level")
plt.ylabel('count')
plt.show()
```

```python
# Ploting hist plot

sns.histplot(New_Dataset['left'] , kde=True)
```

```python
# Printing kurtosis and skewness of data

print(f"Kurtosis : {New_Dataset['left'].kurt()}")
print(f"Skewness : {New_Dataset['left'].skew()}")
```

```python
# Ploting pairplot for the data

sns.pairplot(New_Dataset)
```

```python
# Ploting scatter plot

sns.scatterplot(x="timeSpent.company", y="numberOfProjects", hue="left", data=New_Dataset , palette="Dark2", s=80)
plt.title("Relationship between no. of project , time spent in company and Left")
plt.show()
```

```python
# Correlation of data

corrmat = New_Dataset.corr()
hm = sns.heatmap(corrmat, cbar=True, annot=True, square=True,fmt='.2f',annot_kws={'size': 10},
                 yticklabels=New_Dataset.columns,
                 xticklabels=New_Dataset.columns,
                 cmap="Spectral_r")
plt.show()
```
